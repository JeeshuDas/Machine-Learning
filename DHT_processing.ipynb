```python
import time
import random
import numpy as np
import h5py
from tensorflow.keras.models import load_model
from tensorflow.keras.optimizers import Adam
```

```python
model = load_model('DHT.keras')
```

```python
time_steps = 10
number_of_features = 4
```

```python
def get_real_time_data():
    # Simulating heart rate data with the correct number of features
    return np.array([[random.randint(60, 100) for _ in range(number_of_features)] * time_steps])
```

```python
while True:
    real_time_data = get_real_time_data()
    real_time_data = np.reshape(real_time_data, (1, time_steps, number_of_features))  # Reshape data
    prediction = model.predict(real_time_data)
    print(f"Predicted heart rate: {prediction[0][0]}")
    time.sleep(1)
```
